```python
# import modules
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# send request
url = 'https://en.wikipedia.org/wiki/List_of_automobile_manufacturers'
page = requests.get(url)
page_text = page.text
```

```python
# parsing data
soup = BeautifulSoup(page_text, 'html.parser')
manu_name_elem = soup.find_all('li')
manu_names = []
for elem in manu_name_elem:
    manu_names.append(elem.text)

manu_names
```

```python
# get the index of 'Vinaxuki' for finalized list 
index = manu_names.index('Vinaxuki')
manu_names = manu_names[1:index+1]
manu_names
```

```python
# remove the parenthesis and its content
import re
manu_names = [re.sub(r'\(.*\)|\[.*\]', '', name) for name in manu_names]
```

```python
# check the result
print(set(manu_names))
```

```python
# create dataframe and save it as csv file
df = pd.DataFrame(manu_names, columns=['name_of_manufacturers'])
df.to_csv('manufacturer.csv', index=False)
```
